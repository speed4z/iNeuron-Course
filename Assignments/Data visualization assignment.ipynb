```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets

%matplotlib notebook
```

```python
from sklearn.datasets import load_iris

data = load_iris()
iris = pd.DataFrame(data= np.c_[data['data'], data['target']],columns= data['feature_names'] + ['target'])

iris.head()
```

```python
ax = plt.axes(projection='3d')


# Data for three-dimensional scattered points
xdata = iris['sepal length (cm)']
xlabel=('sepal length (cm)')

ydata = iris['sepal width (cm)']
ylabel=('sepal width (cm)')

zdata = iris['petal length (cm)']
zlabel=('petal length (cm)')

ax.scatter3D(xdata, ydata, zdata, c=iris['petal width (cm)']);

#Here fourth parameter petal width is passed to give various colours
```
